```python
silver_df = spark.read.table("sales_analytics.silver.sales_clean")
```

```python
from pyspark.sql.functions import sum, round
```

```python
region_sales = silver_df.groupBy("Region") \
    .agg(
        round(sum("Final_Revenue"), 2)
        .alias("Total_Sales")
    )
```

```python
region_sales.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales_analytics.gold.region_sales")
```

```python
category_sales = silver_df.groupBy(
    "Product_Category"
).agg(
    round(sum("Final_Revenue"), 2)
    .alias("Category_Sales")
)
```

```python
category_sales.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales_analytics.gold.category_sales")
```

```python
channel_sales = silver_df.groupBy(
    "Sales_Channel"
).agg(
    round(sum("Final_Revenue"), 2)
    .alias("Channel_Sales")
)
```

```python
channel_sales.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales_analytics.gold.channel_sales")
```

```python
from pyspark.sql.functions import month
```

```python
monthly_sales = silver_df.groupBy(
    month("Sale_Date").alias("Month")
).agg(
    round(sum("Final_Revenue"), 2)
    .alias("Monthly_Sales")
)
```

```python
monthly_sales.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales_analytics.gold.monthly_sales")
```

# Total Revenue

```python
%sql
SELECT 
ROUND(SUM(Final_Revenue), 2) AS Total_Revenue
FROM sales_analytics.silver.sales_clean;
```

# Revenue by Region

```python
%sql
SELECT 
Region,
ROUND(SUM(Final_Revenue), 2) AS Revenue
FROM sales_analytics.silver.sales_clean
GROUP BY Region
ORDER BY Revenue DESC;
```

# Revenue by Product Category

```python
%sql
SELECT 
Product_Category,
ROUND(SUM(Final_Revenue), 2) AS Revenue
FROM sales_analytics.silver.sales_clean
GROUP BY Product_Category
ORDER BY Revenue DESC;
```

# Revenue by Sales Channel

```python
%sql
SELECT 
Sales_Channel,
ROUND(SUM(Final_Revenue), 2) AS Revenue
FROM sales_analytics.silver.sales_clean
GROUP BY Sales_Channel;
```

# Monthly Revenue **Trend**

```python
%sql
SELECT 
Month,
ROUND(SUM(Final_Revenue), 2) AS Monthly_Revenue
FROM sales_analytics.silver.sales_clean
GROUP BY Month
ORDER BY Month;
```
