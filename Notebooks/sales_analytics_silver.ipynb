```python
from pyspark.sql.functions import *
```

```python
df = spark.read.table(
    "sales_analytics.bronze.sales_raw"
)
```

```python
silver_df = df.dropna()
```

```python
silver_df = silver_df.withColumn(
    "Sale_Date",
    to_date(col("Sale_Date"), "yyyy-MM-dd")
)
```

```python
silver_df = silver_df.withColumn(
    "Revenue",
    col("Sales_Amount") * col("Quantity_Sold")
)
```

```python
silver_df = silver_df.withColumn(
    "Final_Revenue",
    col("Revenue") * (1 - col("Discount"))
)
```

```python
from pyspark.sql.functions import month, col

silver_df = silver_df.withColumn(
    "Month",
    month(col("Sale_Date"))
)
```

```python
display(silver_df)
```

```python
silver_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("sales_analytics.silver.sales_clean")
```
